```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate example data for poor fit line
np.random.seed(0)
X_poor = 2 * np.random.rand(50, 1)  # Random X values
Y_poor = 4 + 3 * X_poor + np.random.randn(50, 1) * 3  # Y = 4 + 3*X + noise (with larger noise)

# Generate example data for best fit line
X_best = 2 * np.random.rand(50, 1)  # Random X values
Y_best = 4 + 3 * X_best + np.random.randn(50, 1)  # Y = 4 + 3*X + noise

# Fit linear regression models
model_poor = LinearRegression()
model_poor.fit(X_poor, Y_poor)

model_best = LinearRegression()
model_best.fit(X_best, Y_best)

# Make predictions
Y_pred_poor = model_poor.predict(X_poor)
Y_pred_best = model_best.predict(X_best)

# Plotting both lines on the same plot
plt.figure(figsize=(8, 6))

# Plot poor fit line
plt.scatter(X_poor, Y_poor, color='blue', label='Poor Fit Data')
plt.plot(X_poor, Y_pred_poor, color='red', linestyle='-', label='Poor Fit Line')

# Plot best fit line
plt.scatter(X_best, Y_best, color='green', label='Best Fit Data')
plt.plot(X_best, Y_pred_best, color='orange', linestyle='-', label='Best Fit Line')

# Labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Poor Fit vs Best Fit Line')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
